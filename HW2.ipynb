```python
# Задачка N2

import os # импортируем модуль
os.chdir(r'C:\Users\user\Desktop\python-dh-hw\poems')
print(os.getcwd()) # печатаем текущую директорию
file_count = 0
poems_str = ''
for file in os.listdir(): 
    print(file) # печатаем название всех файлов
    if file.find('.') >= 0:
        file_count += 1
        with open(file, 'r') as cur_poem: # открываем файл с кодировкой для чтения
            cur_poem_str = cur_poem.read()
        poems_str += cur_poem_str + '\n'
print(file_count) # получаем количество файлов 
with open('poems.txt', 'w') as poems: # создаем и открываем новый файл poems.txt с кодировкой для добавления содержимого каждого файла
    poems.write(poems_str)
```

```python
# Задачка N3

text = '''Несчастная кошка порезала лапу- 
Сидит, и ни шагу не может ступить.
Скорей, чтобы вылечить кошкину лапу
Воздушные шарики надо купить!

И сразу столпился народ на дороге-
Шумит, и кричит, и на кошку глядит.
А кошка отчасти идет по дороге,
Отчасти по воздуху плавно летит!'''

print(len(text)) # считаем количество символов в тексте
cur_text = [word[:-1] if not word[-1].isalpha() else word for word in text.split()]
print(cur_text) # получаем список слов
cur_text = ' '.join(cur_text) # объединяем слова из получившегося списка в строку через пробел
print(cur_text)
cur_text = [word.lower() if word[0].isupper() else word for word in cur_text.split()] # получаем список из всех слов текста, написанных с маленькой буквы
cur_text_low = cur_text
print(cur_text) 
print(len(cur_text)) # считаем количество слов в тексте
print(len(set(cur_text))) # считаем количество уникальных слов 
print(cur_text[-10:]) # печатаем последние 10 слов
# соединяем все слова из полученного списка в строку через знак переноса строки
cur_text = '\n'.join([word[:-1] if not word[-1].isalpha() else word for word in text.split()])
print(cur_text)
print(cur_text[20:31]) # печатаем с 20 по 30 символ строки (включительно)
# разбиваем текст на биграммы и записываем их в переменную
cur_text = cur_text_low
bigrams = [(cur_text[i], cur_text[i + 1]) for i in range(len(cur_text) - 1)]
d = dict()
for bigram in bigrams:
    if bigram not in d:
        d[bigram] = 0
    d[bigram] += 1
print(d) # считаем частотность каждой биграммы
```


## Задание 2

Очистите лемматизированный текст от стоп-слов и посчитайте ipm для оставшихся. Выведите 20 самых частотных по этому параметру слов.



```python
import os # импортируем модуль
os.chdir(r'C:\Users\user\Desktop\python-dh-hw\poems')
print(os.getcwd()) # печатаем текущую директорию
file_count = 0
poems_str = ''
for file in os.listdir(): # смотрим список всех файлов и папок внутри нее
    print(file) # получаем названия всех файлов в директории
    if file.find('.') >= 0:
        file_count += 1
        with open(file, 'r') as cur_poem: # открываем файл с кодировкой для чтения
            cur_poem_str = cur_poem.read()
        poems_str += cur_poem_str + '\n'
print(file_count) # получаем количество файлов в директории
with open('poems.txt', 'w') as poems: # создаем и открываем новый файл poems.txt с кодировкой для добавления содержимого каждого файла
    poems.write(poems_str) # добавляем содержимый текст в файл
```

Сделайте полный морфологический разбор исходного текста. Напишите регулярное выражение, которое будет извлекать из тега только часть речи.
Пройдитесь циклом по списку с разборами, который выдал `pymystem3`, извлекая из каждого разбора форму слова и его часть речи и записывая их в новый словарь (форма -- ключ, часть речи -- значение).
Посчитайте абсолютную частоту для всех частей речи, а затем относительнную частоту (абсолютная частота / длина текста в словах).

```python
text = '''Несчастная кошка порезала лапу- 
Сидит, и ни шагу не может ступить.
Скорей, чтобы вылечить кошкину лапу
Воздушные шарики надо купить!

И сразу столпился народ на дороге-
Шумит, и кричит, и на кошку глядит.
А кошка отчасти идет по дороге,
Отчасти по воздуху плавно летит!''' # сохраняем текст в переменную
print(len(text)) # получаем длину текста в символах
cur_text = [word[:-1] if not word[-1].isalpha() else word for word in text.split()] # очищаем текст от пунктуаций
print(cur_text) # получаем список слов без пунктуаций
cur_text = ' '.join(cur_text) # объединяем слова из получившегося списка в строку через пробел
print(cur_text)
cur_text = [word.lower() if word[0].isupper() else word for word in cur_text.split()] # проверяем, написано ли слово с большой буквы, если да, то приводим его к нижнему регистру
cur_text_low = cur_text
print(cur_text) # получаем список из всех слов текста, написанных с маленькой буквы
print(len(cur_text)) # считаем количество слов в тексте
print(len(set(cur_text))) # считаем количество уникальных слов 
print(cur_text[-10:]) # печатаем последние 10 слов
cur_text = '\n'.join([word[:-1] if not word[-1].isalpha() else word for word in text.split()]) # соединяем все слова из полученного списка в строку через знак переноса строки
print(cur_text)
print(cur_text[19:31]) # печатаем с 20 по 30 символ строки (включительно)
cur_text = cur_text_low
bigrams = [(cur_text[i], cur_text[i + 1]) for i in range(len(cur_text) - 1)] # проходимся циклом по тексту, разделяя его на биграммы в виде списка кортежей
d = dict()
for bigram in bigrams: # считаем частотность каждой биграммы
    if bigram not in d:
        d[bigram] = 0
    d[bigram] += 1
print(d)
```
